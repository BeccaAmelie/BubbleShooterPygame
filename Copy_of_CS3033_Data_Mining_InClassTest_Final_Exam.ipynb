## Data Preparation
Prepare the dataset between features and labels.

```python
import pandas as pd

# Load the data
df = pd.read_csv('https://github.com/ncrossberg/Datasets/blob/d2ffbf2e72e1842af52b25c37a5458713b3eb461/Final_Dataset_CS3033.csv?raw=true', index_col= 0)

df.head()
```

```python
import numpy as np

# Replace NaN values with 0
df.fillna(0, inplace=True)
```

```python
df.dtypes
```

```python
df['Grade'].value_counts()
```

**T1 - Split your dataset between features (X) and labels (y).**


```python
# X is all columns but the grade, y is just the Grade
X = df.drop("Grade", axis = 1)
y = df["Grade"]
```

**T2 - Split the dataset into 80% training set and 20% test set. Do it in a stratified manner and set the random seed to 42 for reproducibility.**

```python
# split the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## Machine Learning Pipeline



**T3 - Create a basic pipeline that includes**
*   PCA
*   KNN Classifier



```python
# import libraries
from sklearn import neighbors
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
#create pipeline
pipe = Pipeline(steps=[('pca', PCA(8)), ('classifier',neighbors.KNeighborsClassifier(5))])
```

**T4 - Train the pipeline on the training set and evaluate it on the test set.**

```python
# train the pipeline and print accuracy score
pipe.fit(X_train,y_train)
print('Test Accuracy:', pipe.score(X_test, y_test))
```

**Q1 - Is the accuracy you obtained higher than that of a baseline classifier?**





Since there are three different grades: 1,2,3, random classification would have an accuracy of 33%, hence this 58% is almost twice as good.

**T5 - Change the KNN with a Decision Tree Regressor.**

```python
# import decision tree
from sklearn import tree
# create descision tree regressor pipeline
pipe = Pipeline(steps=[('pca', PCA(8)), ('classifier',tree.DecisionTreeRegressor())])
# pipe = Pipeline(steps=[('pca', PCA(8)), ('classifier',tree.DecisionTreeRegressor(random_state = 0))])
```
